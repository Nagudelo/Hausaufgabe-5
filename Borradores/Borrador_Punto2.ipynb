```python
#Estas son las librerías a importar
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

```python
#cargamos el archivo a un arreglo
data = genfromtxt('sparse_sample_monthrg.dat')
```

```python
#inicializamos arreglos nuevos para guardar los datos de interes del archivo
x=arange(0,498)
meses=[]
datos=[]
```

```python
#En unos de los arreglos creados correspondiente a meses ingresaremos los datos que los meses q poseemos.
#Como no tenemos datos de todos los meses de todos los años, este arreglo no tendrá una distancia igual en tiempo entre sus elementos.
#Sin embargo, para poder graficar los datos, le asignamos a cada mes un numero que corresponde a su posicion dentro de una secuancia 
#de meses en donde enero de 1610 es el mes numero 1 y diciembre de 1995 es el mes 4632. 
#Al mismo tiempo, quitamos los meses en los cuales no hay registro de manchas.

for i in x:
    if data[i,3]!= -99:
        posicion=(((data[i,0]-data[0,0])*12)+data[i,1])
        meses.append(posicion)
        datos.append(data[i,3])
tamano=size(meses)
```

```python
#Scatter de los datos. Manchas vs. Meses
scatter(meses, datos)
```

```python
#Utilizando los arreglos "meses" y "datos" realizamos las interpolaciones costantes, lineales y cubicas. Para cumplir con este objetivo
#utilizamos la libreria 'interp1d', la cual importamos desde scipy.interpolate. 
from scipy.interpolate import interp1d
interp  = interp1d(meses, datos,kind=0)
interp1 = interp1d(meses, datos,kind=1)
interp3 = interp1d(meses, datos,kind=3)
```

```python
#Arreglo de datos para evaluar nuestras interpolaciones. Los limites son los mismos que los usados en las interpolaciones
n=linspace(34,4625,5000)
nt=size(n)
```

```python
#grafica de la interpolacion constante
scatter(meses, datos)
plot(n,interp(n))
xlim([0,1500])
```

```python
#grafica de la interpolacion lineal
scatter(meses, datos)
plot(n,interp1(n))
xlim([0,1500])
```

```python
#grafica de la interpolacion cubica
scatter(meses, datos)
plot(n,interp3(n))
xlim([0,1500])
```

```python
#graficas de las interpolaciones juntas
plot(n,interp(n))
plot(n,interp1(n))
plot(n,interp3(n))
xlim([0,2000])
```

```python
#se aprecian las diferentes interpolaciones en a menor escala
plot(n,interp(n))
plot(n,interp1(n))
plot(n,interp3(n))
xlim([0,800])
```

```python
#Teniendo nuestras funciones de interpolacion, procederemos a realizarles la transformada de fourier. Para cumplir este objetivo
#importaremos las librerias 'fft' y 'fftfreq' desde scipy.fftpack. Las funciones 'fft(función)' nos permitirá obtener un arreglo de 
#datos los cuales corresponderan a los coeficientes de Fourier. Por otro lado la funcion 'fftfreq(tamaño, intervalo de tiempo)' nos
#permitirá recuperar las frecuencias para ser utilizadas como variable independiente. 
from scipy.fftpack import fft, fftfreq

fourier=fft(interp(n))
normal=fourier/nt
fourier1=fft(interp1(n))
normal1=fourier1/nt
fourier3=fft(interp3(n))
normal3=fourier3/nt
frequ = fftfreq(nt, 1) # Recuperamos las frecuencias
```
